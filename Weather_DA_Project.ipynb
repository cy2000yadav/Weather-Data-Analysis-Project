# **1) Line Plot for Time Series:**

## Visualize how temperature, dew point temperature, humidity, etc., change over time.

```python
import matplotlib.pyplot as plt

plt.plot(data['Date/Time'], data['Temp_C'], label='Temperature (C)')
plt.plot(data['Date/Time'], data['Dew Point Temp_C'], label='Dew Point Temperature (C)')
plt.xlabel('Date/Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## **2) Histogram for Temperature:**

**To understand the distribution of temperature values.**

```python
plt.hist(data['Temp_C'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

## **3) Scatter Plot for Relationships:**

## Explore relationships between variables.

```python
plt.scatter(data['Temp_C'], data['Rel Hum_%'] , alpha = 0.5)
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

## **4 ) Box Plot for Wind Speed:**

Visualizing the distribution of wind speed.

```python
import seaborn as sns

sns.boxplot(x=data['Wind Speed_km/h'])
plt.xlabel('Wind Speed (km/h)')
plt.show()
```

# **5) Bar Plot for Weather Conditions:**

## Display the count of each weather condition.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=data['Weather Condition'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.show()
```
